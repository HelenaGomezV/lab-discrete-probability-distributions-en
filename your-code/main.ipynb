Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
# Given probabilities
p = 0.6  # probability of apple
q = 0.4  # probability of orange

# 1. Probability that the first 5 fruits are all apples
prob_first_5_apples = p ** 5

# 2. Probability that the first 5 fruits are apples and the next 15 are oranges
prob_5_apples_then_15_oranges = (p ** 5) * (q ** 15)

# Print results
print("Probability first 5 fruits are all apples:", prob_first_5_apples)
print("Probability first 5 apples and next 15 oranges:", prob_5_apples_then_15_oranges)
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
import math

# Given probabilities
p = 0.6  # probability of apple
q = 0.4  # probability of orange

# Parameters
n = 20   # total fruits
k = 5    # number of apples

# Binomial probability
prob_5_apples = math.comb(n, k) * (p ** k) * (q ** (n - k))

# Print result
print("Probability of getting exactly 5 apples and 15 oranges:", prob_5_apples)
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here
from scipy.stats import binom

# Parameters
n = 20      # number of trials
p = 0.6     # probability of apple

# CDF: P(X < 5) = P(X <= 4)
prob_less_than_5_apples = binom.cdf(4, n, p)

print("Probability that fewer than 5 fruits are apples:", prob_less_than_5_apples)
````

<!-- **Plot the PDF (probability distributiom function) for the above binomial distribution.** -->

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 20
p = 0.6

# Range of x values (0 to n)
x = list(range(0, n + 1))

# PMF values
pmf = binom.pmf(x, n, p)

# Plot PMF
plt.figure()
plt.stem(x, pmf)
plt.xlabel("Number of apples (x)")
plt.ylabel("P(X = x)")
plt.title("Binomial PMF (n = 20, p = 0.6)")
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# your code here 
import math

# Given values
lam = 2.3   # average goals per match
k = 5       # number of goals

# Poisson probability
prob_5_goals = (math.exp(-lam) * (lam ** k)) / math.factorial(k)

print("Probability that exactly 5 goals are scored:", prob_5_goals)
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 
import matplotlib.pyplot as plt
import math

# Poisson parameter
lam = 2.3  # average number of goals

# Range of goals from 0 to 10
x = list(range(0, 11))

# Poisson PMF values
pmf = [(math.exp(-lam) * (lam ** k)) / math.factorial(k) for k in x]

# Plot the Poisson distribution
plt.figure()
plt.stem(x, pmf)
plt.xlabel("Number of goals")
plt.ylabel("P(X = k)")
plt.title("Poisson Distribution for Goals Scored ( = 2.3)")
plt.show()
````
